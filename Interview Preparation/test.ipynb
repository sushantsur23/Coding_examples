```python
str1 = 'bat'
str2 = 'tab'
d = {}
if len(str1)!=len(str2):
    print(":Not anagram")
else:
    for i in str1:
        if i in d.keys():
            d[i]+=1
        else:
            d[i]=1
    print("Initial dict",d)
    for j in str2:
        if j in d.keys():
            d[j]-=1
        if d[j] ==0:
            del d[j]

    if not d:
        print("Anagram")
    else:
        print("Not Anagram")

```

```python
#Anagram

str1 = "tab"
str2 = "bat"
d= {}

if len(str1)!=len(str2):

    print("not Anagram")
else:
    for i in str1:
        if i in d.keys():
            d[i]+=1
        else:
            d[i]=1
    print("Old Dict is ",d)
    for j in str2:
        if j in d.keys():
            d[j]-=1
        if d[j]==0:
            del d[j]
    print("New dict is",d)
    if len(d)!=0:
        print("Not Anagram")
    else:
        print("Anagram")
```

```python
#reverse a string with generator

def reverse(st: str) -> str:
    length = len(st)
    for i in range (length-1,-1,-1):
        yield st[i]

for char in reverse("Hello"):
    print(char)
```

```python
def fact(n):
    if n<=1:
        return 1
    else:
        return n*fact(n-1)
    

print("Factorial is ",fact(10))
```
